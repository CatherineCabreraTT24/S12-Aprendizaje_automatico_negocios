# PARTE 1 — Métricas financieras

Vamos a resolver el caso del hook. Usamos datos de 3 meses de un e-commerce de ropa deportiva.

La cadena de métricas es siempre la misma:


Ingresos 
* \- Costo de bienes vendidos  = Beneficio bruto      (¿el producto vale la pena?)
* \- Gastos operativos = Beneficio operativo   (¿el negocio funciona?)
* \- Impuestos y préstamos = Beneficio neto        (¿qué queda para los dueños?)

````python
finanzas = pd.DataFrame({
    'mes':                    ['Enero',  'Febrero', 'Marzo'],
    'ingresos':               [120_000,  145_000,   98_000],
    'costo_bienes_vendidos':  [ 72_000,   87_000,   62_000],
    'gastos_operativos':      [ 35_000,   38_000,   40_000],
    'impuestos_prestamos':    [  4_500,    5_200,    3_100],
    'inversion_marketing':    [ 15_000,   20_000,   25_000]
})

# Calculamos todas las métricas encadenadas
finanzas['beneficio_bruto']     = finanzas['ingresos'] - finanzas['costo_bienes_vendidos']
finanzas['margen_bruto']        = finanzas['beneficio_bruto'] / finanzas['ingresos']
finanzas['beneficio_operativo'] = finanzas['beneficio_bruto'] - finanzas['gastos_operativos']
finanzas['margen_operativo']    = finanzas['beneficio_operativo'] / finanzas['ingresos']
finanzas['beneficio_neto']      = finanzas['beneficio_operativo'] - finanzas['impuestos_prestamos']
finanzas['roi_marketing']       = (finanzas['beneficio_neto'] - finanzas['inversion_marketing']) / finanzas['inversion_marketing']

print('=== RESUMEN FINANCIERO ===')
for _, row in finanzas.iterrows():
    print(f"\n📅 {row['mes']}")
    print(f"  Ingresos:             ${row['ingresos']:>10,.0f}")
    print(f"  Beneficio bruto:      ${row['beneficio_bruto']:>10,.0f}  ({row['margen_bruto']:.1%})")
    print(f"  Beneficio operativo:  ${row['beneficio_operativo']:>10,.0f}  ({row['margen_operativo']:.1%})")
    print(f"  Beneficio neto:       ${row['beneficio_neto']:>10,.0f}")
    emoji = '✅' if row['roi_marketing'] > 0 else '❌'
    print(f"  ROI marketing:        {row['roi_marketing']:>10.1%}  {emoji}")
````

````python
# Visualización de márgenes por mes
fig, ax = plt.subplots(figsize=(10, 5))

x = np.arange(3)
w = 0.25
ax.bar(x - w,   finanzas['margen_bruto'],      w, label='Margen bruto',      color='#4C9BE8', alpha=0.85)
ax.bar(x,       finanzas['margen_operativo'],   w, label='Margen operativo',  color='#F4845F', alpha=0.85)
ax.bar(x + w,   finanzas['roi_marketing'],      w, label='ROI marketing',     color='#65BD82', alpha=0.85)

ax.axhline(0, color='black', linewidth=0.8, linestyle='--')
ax.set_xticks(x)
ax.set_xticklabels(finanzas['mes'], fontsize=12)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
ax.set_title('Márgenes y ROI por mes', fontsize=13, fontweight='bold')
ax.legend(fontsize=11)
ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()

print('💬 Marzo tiene el ROI más bajo. ¿Cancelarías el presupuesto de marketing?')
print('   Escribe tu argumento en el chat antes de seguir...')
````

# PARTE 2 — Embudo de conversión

Los ingresos dependen de cuántos usuarios completan el flujo de compra. El embudo muestra exactamente dónde se pierden.

```
Visitantes → Ven el producto → Lo agregan → Pagan → Compran
   100%           80%             30%         10%      1%
                                               ↑
```

````python
embudo = pd.DataFrame({
    'etapa':    ['Página principal', 'Página de producto', 'Añadido al carrito', 'Página de pago', 'Compra completada'],
    'usuarios': [85_000, 62_000, 21_000, 9_500, 3_200]
})

total = embudo['usuarios'].iloc[0]
embudo['conversion_total']     = embudo['usuarios'] / total
embudo['conversion_siguiente'] = embudo['usuarios'] / embudo['usuarios'].shift(1)
embudo['conversion_siguiente'] = embudo['conversion_siguiente'].fillna(1.0)

print(embudo.to_string(index=False,
    formatters={
        'usuarios':             '{:,.0f}'.format,
        'conversion_total':     '{:.1%}'.format,
        'conversion_siguiente': '{:.1%}'.format
    }))
````

````python
fig, ax = plt.subplots(figsize=(10, 5))

colores = ['#4C9BE8', '#5EAAF0', '#70B9F8', '#82C8FF', '#F4845F']
y_pos   = range(len(embudo))
bars    = ax.barh(list(y_pos), embudo['usuarios'], color=colores, height=0.55, edgecolor='white')

for bar, row in zip(bars, embudo.itertuples()):
    ax.text(bar.get_width() + 600, bar.get_y() + bar.get_height()/2,
            f'{row.usuarios:,.0f}  ({row.conversion_total:.1%})',
            va='center', fontsize=11)

ax.set_yticks(list(y_pos))
ax.set_yticklabels(embudo['etapa'], fontsize=11)
ax.invert_yaxis()
ax.set_xlabel('Usuarios', fontsize=12)
ax.set_title('Embudo de conversión — E-commerce deportivo', fontsize=13, fontweight='bold')
ax.set_xlim(0, 100_000)
ax.xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/1000:.0f}K'))
ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()
````

### 🐛 Bug hunt — ¿Cuál es el error lógico?

El siguiente código busca la **etapa con mayor caída de usuarios**, pero el resultado es incorrecto.
Escribe en el chat qué está mal antes de ejecutar la corrección.

````python
# 🐛 CÓDIGO CON BUG
embudo['usuarios_perdidos'] = embudo['usuarios'].diff()
etapa_critica = embudo.loc[embudo['usuarios_perdidos'].idxmax()]

print('Etapa crítica (mayor caída):')
print(f"  {etapa_critica['etapa']} — perdidos: {etapa_critica['usuarios_perdidos']:,.0f}")
````

````python
# ✅ VERSIÓN CORREGIDA
# Bug: .diff() da valores NEGATIVOS cuando hay caída.
# .idxmax() encontraba la caída MÁS PEQUEÑA (menos negativa), no la mayor.
# Solución: usar .abs() para trabajar con valores positivos.

embudo['usuarios_perdidos'] = embudo['usuarios'].diff().abs()
etapa_critica = embudo.loc[embudo['usuarios_perdidos'].idxmax()]

print('✅ Etapa con mayor caída de usuarios:')
print(f"   → {etapa_critica['etapa']}: {etapa_critica['usuarios_perdidos']:,.0f} usuarios perdidos\n")

print('📋 Pérdidas por etapa:')
for _, row in embudo.dropna(subset=['usuarios_perdidos']).iterrows():
    barra = '█' * int(row['usuarios_perdidos'] / 2000)
    print(f"  {row['etapa']:<25} {barra} {row['usuarios_perdidos']:>7,.0f}")
````

# PARTE 3 — Métricas online vs offline

La empresa tiene un modelo de recomendaciones. ¿Cómo lo evaluamos?

| Métrica | ¿Dónde se calcula? | Ejemplo |
|---|---|---|
| **Offline** | Datos históricos, antes de lanzar | Exactitud, precisión, recall, F1 |
| **Online** | En producción, con usuarios reales | Precio medio de ítems recomendados que se compraron |

La métrica online es la que realmente importa al negocio, pero **no la puedes calcular hasta que el modelo está en producción**. Por eso primero validas offline y luego mides online.

````python
np.random.seed(42)
n = 500

sin_recom = np.random.normal(loc=45, scale=18, size=n).clip(0)
con_recom = np.random.normal(loc=58, scale=20, size=n).clip(0)

print('=== MÉTRICAS OFFLINE (calculadas sobre histórico) ===')
print(f'  Sin recomendaciones — Ticket promedio: ${sin_recom.mean():.2f}')
print(f'  Con recomendaciones — Ticket promedio: ${con_recom.mean():.2f}')
diff = con_recom.mean() - sin_recom.mean()
print(f'  Diferencia estimada: +${diff:.2f}  ({diff/sin_recom.mean():.1%})')

print('\n=== MÉTRICA ONLINE (requiere producción) ===')
print('  Precio medio de ítems añadidos por recomendación: ⚠️ no calculable offline')
print('  → Solo existe cuando usuarios reales interactúan con el modelo en vivo')
````

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

for ax, datos, color, c_linea, titulo in zip(
    axes,
    [sin_recom, con_recom],
    ['#4C9BE8', '#F4845F'],
    ['navy', 'darkred'],
    ['Sin recomendaciones', 'Con recomendaciones']
):
    ax.hist(datos, bins=35, color=color, alpha=0.8, edgecolor='white')
    ax.axvline(datos.mean(), color=c_linea, linestyle='--', linewidth=2,
               label=f'Media: ${datos.mean():.1f}')
    ax.set_title(titulo, fontweight='bold')
    ax.set_xlabel('Ticket promedio ($)')
    ax.legend(fontsize=11)
    ax.spines[['top', 'right']].set_visible(False)

plt.suptitle('Distribución del ticket — Offline (histórico simulado)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
````

# 📝 Resumen — ¿Qué aprendimos hoy?

| Métrica | Fórmula | ¿Para qué sirve? |
|---|---|---|
| **Beneficio bruto** | Ingresos − COGS | Salud básica del producto |
| **Beneficio operativo** | B. bruto − Gastos operativos | Gestión diaria |
| **Beneficio neto** | B. operativo − Impuestos/préstamos | Junta anual |
| **ROI** | (B. neto − Inversión) / Inversión | Decisiones de inversión |
| **Conversión** | Usuarios que completan / Total | Eficacia del embudo |
| **Métrica offline** | Calculada con datos históricos | Evaluar modelo antes de lanzar |
| **Métrica online** | Con usuarios reales en producción | Evaluar impacto real |

> CEO: "Tenemos $500K en ingresos, somos rentables."

- Beneficio bruto: $500K − $420K = **$80,000** ✅
